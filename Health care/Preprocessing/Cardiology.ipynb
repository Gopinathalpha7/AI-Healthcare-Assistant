# Problem Statement:
--------------------
Cardiovascular diseases (CVD) are a major global health concern, responsible for millions of deaths annually. Early detection of individuals at risk is crucial for preventing severe health outcomes. This project aims to develop a machine learning model to predict the presence or absence of cardiovascular disease based on a combination of patient demographics, medical examination results, and lifestyle factors, all captured during routine medical check-ups.

### Key Objectives:
1. Identify key factors (objective, examination, and subjective features) contributing to cardiovascular disease risk.
2. Develop and train a predictive model using the available dataset to classify patients as either at risk or not at risk for cardiovascular disease.
3. Evaluate model performance using appropriate metrics (accuracy, precision, recall, F1-score, etc.) to ensure reliability.
4. Provide insights into the most important features that influence the model's predictions.

### Dataset link: [Cardiovascular Disease Dataset](https://www.kaggle.com/datasets/sulianova/cardiovascular-disease-dataset/data)

### Approach:
1. **Data Preprocessing**: Clean and preprocess the dataset, including handling missing values, normalizing/standardizing numerical features, and encoding categorical variables.
2. **Exploratory Data Analysis (EDA)**: Perform EDA to understand the distribution of features and detect potential correlations with the target variable (cardiovascular disease).
3. **Feature Selection**: Identify the most important features using techniques like correlation analysis or feature importance from machine learning models.
4. **Model Development**: Implement machine learning algorithms such as Logistic Regression, Decision Trees, Random Forest, or Gradient Boosting to predict cardiovascular disease.
5. **Model Evaluation**: Use cross-validation and performance metrics like accuracy, precision, recall, F1-score, and ROC-AUC to evaluate the models.
6. **Model Interpretation**: Analyze the feature importance or SHAP values to interpret the model's decision-making process.

### Results:
The expected outcome is a well-performing machine learning model capable of predicting the presence or absence of cardiovascular disease with high accuracy and providing insights into the most significant factors contributing to cardiovascular risk. The model can be utilized by healthcare professionals to enhance early diagnosis and preventive measures.

# About Dataset
--------------------

**Data Description**

There are 3 types of input features:

- **Objective**: factual information
- **Examination**: results of medical examination
- **Subjective**: information given by the patient

### Features:

1. **Age** | Objective Feature | `age` | int (days)
2. **Height** | Objective Feature | `height` | int (cm)
3. **Weight** | Objective Feature | `weight` | float (kg)
4. **Gender** | Objective Feature | `gender` | categorical code
    - 1: women
    - 2: men
5. **Systolic blood pressure** | Examination Feature | `ap_hi` | int
6. **Diastolic blood pressure** | Examination Feature | `ap_lo` | int
7. **Cholesterol** | Examination Feature | `cholesterol` | categorical
    - 1: normal
    - 2: above normal
    - 3: well above normal
8. **Glucose** | Examination Feature | `gluc` | categorical
    - 1: normal
    - 2: above normal
    - 3: well above normal
9. **Smoking** | Subjective Feature | `smoke` | binary
10. **Alcohol intake** | Subjective Feature | `alco` | binary
11. **Physical activity** | Subjective Feature | `active` | binary
12. **Presence or absence of cardiovascular disease** | Target Variable | `cardio` | binary

**Note**: All of the dataset values were collected at the time of the medical examination.


Dataset link: https://www.kaggle.com/datasets/sulianova/cardiovascular-disease-dataset/data

### Model Selection.

#### 1. **Logistic Regression**

### Model Building and Fine turning (LGBMClassifier)

```python
# Import libraries
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import cross_val_score
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create an array to store the average accuracy for different n_estimators
n_estimators_range = range(50, 201, 10)  # Test estimators from 50 to 200 with a step of 10
cv_scores = []  # To store cross-validation scores

# Step 2: Iterate over different n_estimators and use cross-validation
for n_estimators in n_estimators_range:
    lgbm = LGBMClassifier(n_estimators=n_estimators, random_state=42)
    scores = cross_val_score(lgbm, X_train, y_train, cv=5, scoring='accuracy')
    cv_scores.append(scores.mean())  # Store the mean of cross-validation scores

# Step 3: Plot the results to find the optimum n_estimators
plt.plot(n_estimators_range, cv_scores, marker='o')
plt.xlabel('Number of Estimators')
plt.ylabel('Cross-Validated Accuracy')
plt.title('Optimum Number of Estimators for LightGBM')
plt.show()

# Step 4: Choose the n_estimators with the highest cross-validation score
optimal_n_estimators = n_estimators_range[np.argmax(cv_scores)]
print(f"Optimal number of estimators: {optimal_n_estimators}")

# Step 5: Train the LightGBM classifier with the optimal n_estimators
lgbm_optimal = LGBMClassifier(n_estimators=optimal_n_estimators, random_state=42)
lgbm_optimal.fit(X_train, y_train)

# Step 6: Make predictions on the test set
y_pred = lgbm_optimal.predict(X_test)

# Step 7: Evaluate accuracy of the model
lgbm_accuracy = accuracy_score(y_test, y_pred) * 100
print(f"LightGBM Accuracy: {lgbm_accuracy:.2f}%")
```

### Model Building and Fine turning DecisionTreeClassifier

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import cross_val_score

# Step 1: Create an array to store the average accuracy for different max_depth values
max_depths = range(1, 21)  # Test depths from 1 to 20
cv_scores = []  # To store cross-validation scores

# Step 2: Iterate over different depths and use cross-validation
for depth in max_depths:
    clf = DecisionTreeClassifier(max_depth=depth, random_state=42)
    scores = cross_val_score(clf, X_train, y_train, cv=5, scoring='accuracy')
    cv_scores.append(scores.mean())  # Store the mean of cross-validation scores

# Step 3: Plot the results to find the optimum depth
plt.plot(max_depths, cv_scores, marker='o')
plt.xlabel('Max Depth')
plt.ylabel('Cross-Validated Accuracy')
plt.title('Optimum Depth for Decision Tree')
plt.show()

# Step 4: Choose the depth with the highest cross-validation score
optimal_depth = max_depths[np.argmax(cv_scores)]
print(f"Optimal depth: {optimal_depth}")

# Step 5: Train the Decision Tree classifier with the optimal depth
clf_optimal = DecisionTreeClassifier(max_depth=optimal_depth, random_state=42)
clf_optimal.fit(X_train, y_train)
```
