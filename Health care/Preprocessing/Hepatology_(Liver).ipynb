# Problem Statement:
--------------------
Liver disease is a critical global health issue, with significant morbidity and mortality rates. Early diagnosis is essential for effective treatment and management, potentially improving patient outcomes and quality of life. This project aims to develop a machine learning model to predict the likelihood of liver disease in patients based on clinical indicators and demographic information collected from medical examinations.

### Key Objectives:
1. Identify key demographic and clinical factors contributing to the risk of liver disease.
2. Develop and train a predictive model using the provided dataset to classify patients as having liver disease or not.
3. Evaluate model performance using metrics such as accuracy, precision, recall, F1-score, etc., to ensure reliability.
4. Provide insights into the most influential features impacting the model's predictions to assist healthcare professionals in diagnosis.

### Dataset link: [Liver Disease Patient Dataset](https://www.kaggle.com/datasets/abhi8923shriv/liver-disease-patient-dataset?select=Liver+Patient+Dataset+%28LPD%29_train.csv)

### Approach:
1. **Data Preprocessing**: Clean and preprocess the dataset by handling missing values, normalizing/standardizing numerical features, and encoding categorical variables.
2. **Exploratory Data Analysis (EDA)**: Perform EDA to understand feature distributions and potential correlations with the target variable (liver disease status).
3. **Feature Selection**: Identify the most relevant features using techniques such as correlation analysis or feature importance from machine learning models.
4. **Model Development**: Implement machine learning algorithms such as Logistic Regression, Decision Trees, Random Forest, or Gradient Boosting to predict liver disease status.
5. **Model Evaluation**: Use cross-validation and performance metrics like accuracy, precision, recall, F1-score, and ROC-AUC to assess the models.
6. **Model Interpretation**: Analyze feature importance to understand the model’s decision-making process and offer interpretable results to assist medical practitioners.

### Results:
The expected outcome is a reliable machine learning model capable of predicting the presence or absence of liver disease with high accuracy. The model will also yield insights into significant clinical factors associated with liver disease, assisting healthcare professionals in early diagnosis and personalized treatment planning.

# About Dataset
--------------------

**Data Description**

The liver disease patient dataset includes various medical and demographic features that could assist in predicting liver disease. The data comprises key clinical indicators and demographic information collected during medical check-ups.

### Features:

1. **Age** | Patient's age | `age` | int
2. **Gender** | Patient's gender | `gender` | categorical
   - 1: male
   - 2: female
3. **Total Bilirubin (TB)** | Blood marker for liver function | `total_bilirubin` | float (mg/dL)
4. **Direct Bilirubin (DB)** | Direct measure of bilirubin | `direct_bilirubin` | float (mg/dL)
5. **Alkaline Phosphotase (Alkphos)** | Enzyme associated with liver function | `alkphos` | float (U/L)
6. **Alamine Aminotransferase (SGPT)** | Enzyme associated with liver health | `sgpt` | float (U/L)
7. **Aspartate Aminotransferase (SGOT)** | Enzyme associated with liver health | `sgot` | float (U/L)
8. **Total Proteins (TP)** | Total protein level | `total_proteins` | float (g/dL)
9. **Albumin (ALB)** | Blood protein | `albumin` | float (g/dL)
10. **Albumin and Globulin Ratio (A/G Ratio)** | Protein ratio | `a_g_ratio` | float

**Target Variable**: Liver Disease Status (`Selector`)
   - 1: Liver Patient
   - 2: Non-Liver Patient

### Model Selection.

#### 1. **Logistic Regression**

```python
# Import library
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Initialize
log_reg = LogisticRegression()

# Fit the model
log_reg.fit(X_train, y_train)

# Predictions
y_pred = log_reg.predict(X_test)

# Accuracy of model in %
logreg_accuracy = accuracy_score(y_test, y_pred) * 100
print(f"Logistic Regression Accuracy: {logreg_accuracy:.2f}%")
```

#### 2. **K-Nearest Neighbors (KNN)**

```python
# Import library
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Initialize
knn = KNeighborsClassifier()

# Fit the model
knn.fit(X_train, y_train)

# Predictions
y_pred = knn.predict(X_test)

# Accuracy of model in %
knn_accuracy = accuracy_score(y_test, y_pred) * 100
print(f"K-Nearest Neighbors Accuracy: {knn_accuracy:.2f}%")
```

#### 3. **Support Vector Machines (SVM)**

```python
# Import library
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Initialize
svm = SVC()

# Fit the model
svm.fit(X_train, y_train)

# Predictions
y_pred = svm.predict(X_test)

# Accuracy of model in %
svm_accuracy = accuracy_score(y_test, y_pred) * 100
print(f"Support Vector Machine Accuracy: {svm_accuracy:.2f}%")
```

#### 4. **Decision Tree Classifier**

### Model Building and Fine turning (Random Forest)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score

# Step 1: Define the range of estimators to test
n_estimators_range = range(10, 210, 10)  # Testing n_estimators from 10 to 200
cv_scores = []  # To store cross-validation scores

# Step 2: Iterate over different numbers of trees (n_estimators)
for n_estimators in n_estimators_range:
    rfc = RandomForestClassifier(n_estimators=n_estimators, random_state=42)
    scores = cross_val_score(rfc, X_train, y_train, cv=5, scoring='accuracy')
    cv_scores.append(np.mean(scores))  # Store the mean cross-validation score

# Step 3: Plot the results to find the optimum number of trees
plt.plot(n_estimators_range, cv_scores, marker='o')
plt.xlabel('Number of Trees (n_estimators)')
plt.ylabel('Cross-Validated Accuracy')
plt.title('Optimum Number of Trees for Random Forest')
plt.show()

# Step 4: Choose the number of trees with the highest cross-validation score
optimal_n_estimators = n_estimators_range[np.argmax(cv_scores)]
print(f"Optimal number of trees: {optimal_n_estimators}")

# Step 5: Train the Random Forest classifier with the optimal number of trees
rfc_optimal = RandomForestClassifier(n_estimators=optimal_n_estimators, random_state=42)
rfc_optimal.fit(X_train, y_train)

# Step 6: Evaluate the model on the test set
y_pred = rfc_optimal.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Test set accuracy: {accuracy*100:.2f}")
```
