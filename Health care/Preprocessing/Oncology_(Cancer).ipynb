# Problem Statement:
--------------------
Cancer is a leading cause of death worldwide, and timely diagnosis is crucial for improving survival rates and treatment outcomes. This project aims to develop a predictive model to assess the likelihood of cancer diagnosis based on various clinical and demographic features provided in the dataset. By identifying patterns and risk factors associated with cancer, the model can aid healthcare providers in early detection and personalized treatment planning.

### Key Objectives:
1. Identify key demographic and clinical factors that correlate with cancer diagnosis.
2. Develop and train a predictive model to classify the likelihood of cancer diagnosis using the provided dataset.
3. Evaluate model performance using metrics such as accuracy, precision, recall, F1-score, and ROC-AUC to ensure reliability.
4. Provide insights into the most influential features impacting the model's predictions, assisting healthcare professionals in better understanding the cancer risk factors.

### Dataset link: [Cancer Diagnosis Prediction Dataset](https://www.kaggle.com/competitions/mubravo/data)

### Approach:
1. **Data Preprocessing**: Handle missing values, normalize/standardize numerical features, and encode categorical variables.
2. **Exploratory Data Analysis (EDA)**: Investigate feature distributions and relationships with the target variable (cancer diagnosis).
3. **Feature Selection**: Identify and retain relevant features using techniques like correlation analysis and feature importance.
4. **Model Development**: Implement machine learning algorithms (e.g., Logistic Regression, Decision Trees, Random Forest, Gradient Boosting) to predict cancer diagnosis.
5. **Model Evaluation**: Use cross-validation and performance metrics (accuracy, precision, recall, F1-score, and ROC-AUC) for assessment.
6. **Model Interpretation**: Examine feature importance to understand the model’s decision-making process and offer interpretability to aid medical practitioners.

### Results:
The project aims to create a robust model capable of predicting cancer diagnosis with high accuracy, providing insights into significant clinical factors related to cancer. This will support healthcare professionals in early detection and tailored treatment strategies.

13. **LymphInfil** | Presence of lymphatic infiltration (ordinal: values from 1 to 3)
14. **ESR1** | ESR1 gene expression level (continuous)

```python
import numpy as np

Numerical_Features = ['Age', 'Time Since Recurrence', 'Tumor Diameter', 'Positive Lymph Nodes', 'ESR1 Gene Expression']

# Create a dictionary to hold filtered data for each feature
filtered_data_dict = {}

for feature in Numerical_Features:
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = np.percentile(data[feature], 25)
    Q3 = np.percentile(data[feature], 75)

    # Calculate IQR
    IQR = Q3 - Q1

    # Define bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Remove outliers and store the filtered data
    filtered_data_dict[feature] = data[feature][(data[feature] >= lower_bound) & (data[feature] <= upper_bound)]

# The filtered data for each feature is now available in filtered_data_dict
```

### Model Building and Fine turning (Random Forest)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score

# Step 1: Define the range of estimators to test
n_estimators_range = range(10, 310, 5)  # Testing n_estimators from 10 to 200
cv_scores = []  # To store cross-validation scores

# Step 2: Iterate over different numbers of trees (n_estimators)
for n_estimators in n_estimators_range:
    rfc = RandomForestClassifier(n_estimators=n_estimators, random_state=42)
    scores = cross_val_score(rfc, X_train, y_train, cv=5, scoring='accuracy')
    cv_scores.append(np.mean(scores))  # Store the mean cross-validation score

# Step 3: Plot the results to find the optimum number of trees
plt.plot(n_estimators_range, cv_scores, marker='o')
plt.xlabel('Number of Trees (n_estimators)')
plt.ylabel('Cross-Validated Accuracy')
plt.title('Optimum Number of Trees for Random Forest')
plt.show()

# Step 4: Choose the number of trees with the highest cross-validation score
optimal_n_estimators = n_estimators_range[np.argmax(cv_scores)]
print(f"Optimal number of trees: {optimal_n_estimators}")

# Step 5: Train the Random Forest classifier with the optimal number of trees
rfc_optimal = RandomForestClassifier(n_estimators=optimal_n_estimators, random_state=42)
rfc_optimal.fit(X_train, y_train)

# Step 6: Evaluate the model on the test set
y_pred = rfc_optimal.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Test set accuracy: {accuracy*100:.2f}")
```
