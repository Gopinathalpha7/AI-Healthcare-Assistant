# Problem Statement:
--------------------
Chronic Kidney Disease (CKD) is a serious health condition that, if left undiagnosed, can lead to end-stage kidney failure and other severe complications. Early diagnosis and management of CKD are critical for reducing morbidity, slowing disease progression, and improving patient outcomes. This project aims to develop a machine learning model to predict the likelihood of CKD in patients based on clinical and demographic indicators collected from medical examinations.

### Key Objectives:
1. Identify key clinical and demographic factors contributing to the risk of CKD.
2. Develop and train a predictive model using the provided dataset to classify patients as having CKD or not.
3. Evaluate model performance using metrics such as accuracy, precision, recall, F1-score, and AUC-ROC to ensure reliability.
4. Provide insights into the most influential features impacting model predictions to aid healthcare professionals in CKD diagnosis and treatment.

### Dataset link: [Chronic Kidney Disease Dataset](https://www.kaggle.com/datasets/akshayksingh/kidney-disease-dataset)

### Approach:
1. **Data Preprocessing**: Clean and preprocess the dataset by handling missing values, normalizing numerical features, and encoding categorical variables.
2. **Exploratory Data Analysis (EDA)**: Perform EDA to understand feature distributions and correlations with CKD status.
3. **Feature Selection**: Identify the most relevant features using techniques such as correlation analysis or feature importance from machine learning models.
4. **Model Development**: Implement machine learning algorithms such as Logistic Regression, Decision Trees, Random Forest, or Gradient Boosting to predict CKD status.
5. **Model Evaluation**: Use cross-validation and performance metrics like accuracy, precision, recall, F1-score, and ROC-AUC to assess model robustness.
6. **Model Interpretation**: Analyze feature importance to understand the model’s decision-making process and provide interpretable results to assist medical practitioners.

### Expected Outcome:
The expected outcome is a reliable machine learning model capable of predicting the presence or absence of CKD with high accuracy. The model will also yield insights into significant clinical factors associated with CKD, supporting healthcare providers in early diagnosis and personalized treatment planning.

15. **Hemoglobin (HEMO)** | Hemoglobin level | `hemo` | float (g/dL)
16. **Packed Cell Volume (PCV)** | Volume of packed red cells | `pcv` | int
17. **White Blood Cell Count (WBCC)** | White blood cell count | `wbcc` | int (cells/cmm)

```python
import numpy as np

Numerical_Features = ['Age', 'BP', 'BGR', 'BU', 'SC', 'SOD', 'POT', 'HEMO', 'PCV', 'WBCC', 'RBCC']

# Create a dictionary to hold filtered data for each feature
filtered_data_dict = {}

for feature in Numerical_Features:
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = np.percentile(data[feature], 25)
    Q3 = np.percentile(data[feature], 75)

    # Calculate IQR
    IQR = Q3 - Q1

    # Define bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Remove outliers and store the filtered data
    filtered_data_dict[feature] = data[feature][(data[feature] >= lower_bound) & (data[feature] <= upper_bound)]

# The filtered data for each feature is now available in filtered_data_dict
```

### Model Building and Fine turning (SVM)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score

# Step 1: Define the range of C values to test
C_range = np.logspace(-3, 2, 10)  # Testing C from 0.001 to 100 on a logarithmic scale
cv_scores = []  # To store cross-validation scores

# Step 2: Iterate over different values of C
for C in C_range:
    svm = SVC(C=C, kernel='linear', random_state=42)
    scores = cross_val_score(svm, X_train, y_train, cv=5, scoring='accuracy')
    cv_scores.append(np.mean(scores))  # Store the mean cross-validation score

# Step 3: Plot the results to find the optimum C value
plt.plot(C_range, cv_scores, marker='o')
plt.xlabel('Regularization Parameter (C)')
plt.ylabel('Cross-Validated Accuracy')
plt.title('Optimum C for SVM')
plt.xscale('log')  # Log scale for C values
plt.show()

# Step 4: Choose the C value with the highest cross-validation score
optimal_C = C_range[np.argmax(cv_scores)]
print(f"Optimal C value: {optimal_C}")

# Step 5: Train the SVM classifier with the optimal C value
svm_optimal = SVC(C=optimal_C, kernel='linear', random_state=42)
svm_optimal.fit(X_train, y_train)

# Step 6: Evaluate the model on the test set
y_pred = svm_optimal.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Test set accuracy: {accuracy*100:.2f}")
```
