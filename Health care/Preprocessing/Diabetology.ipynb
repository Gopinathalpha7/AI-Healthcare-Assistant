# Problem Statement:
--------------------
Diabetes is a significant global health issue, affecting millions of individuals and leading to severe health complications if left unmanaged. Early diagnosis is critical to prevent the progression of the disease. This project aims to develop a machine learning model to predict the likelihood of diabetes in patients based on a combination of demographic information, medical history, and lifestyle factors captured during routine check-ups.

### Key Objectives:
1. Identify key demographic and medical factors contributing to the risk of diabetes.
2. Develop and train a predictive model using the available dataset to classify patients as diabetic or non-diabetic.
3. Evaluate model performance using appropriate metrics (accuracy, precision, recall, F1-score, etc.) to ensure reliability.
4. Provide insights into the most important features influencing the model's predictions.

### Dataset link: [Diabetes Prediction Dataset](https://www.kaggle.com/datasets/iammustafatz/diabetes-prediction-dataset)

### Approach:
1. **Data Preprocessing**: Clean and preprocess the dataset by handling missing values, normalizing/standardizing numerical features, and encoding categorical variables.
2. **Exploratory Data Analysis (EDA)**: Perform EDA to understand the distribution of features and detect potential correlations with the target variable (diabetes).
3. **Feature Selection**: Identify the most relevant features using techniques such as correlation analysis or feature importance from machine learning models.
4. **Model Development**: Implement machine learning algorithms such as Logistic Regression, Decision Trees, Random Forest, or Gradient Boosting to predict diabetes.
5. **Model Evaluation**: Use cross-validation and performance metrics like accuracy, precision, recall, F1-score, and ROC-AUC to evaluate the models.
6. **Model Interpretation**: Analyze feature importance or SHAP values to interpret the model's decision-making process.

### Results:
The expected outcome is a well-performing machine learning model capable of predicting the presence or absence of diabetes with high accuracy. The model will also provide valuable insights into the most significant factors contributing to diabetes risk, aiding healthcare professionals in early diagnosis and personalized treatment planning.


# About Dataset
--------------------

**Data Description**

The diabetes prediction dataset contains medical and demographic information collected from patients to help predict their diabetes status. The data is categorized into several key features, which include patient demographics, medical history, and lifestyle factors.

### Features:

1. **Age** | Objective Feature | `age` | int
2. **Gender** | Objective Feature | `gender` | categorical
   - 1: male
   - 2: female
3. **Body Mass Index (BMI)** | Examination Feature | `bmi` | float (kg/m²)
4. **Hypertension** | Examination Feature | `hypertension` | binary
   - 0: no
   - 1: yes
5. **Heart Disease** | Examination Feature | `heart_disease` | binary
   - 0: no
   - 1: yes
6. **Smoking History** | Subjective Feature | `smoking_history` | categorical
   - 1: never
   - 2: No Info
   - 3: current
   - 4: former
   - 5: not current
   - 6: ever

7. **HbA1c Level** | Examination Feature | `hba1c_level` | float (%)
8. **Blood Glucose Level** | Examination Feature | `blood_glucose_level` | float (mg/dL)
9. **Diabetes Status** | Target Variable | `diabetes` | binary
   - 0: negative
   - 1: positive

**Note**: The dataset includes patients' medical and lifestyle information that can be used to predict diabetes status, with features collected during routine medical check-ups.

### Dataset link: [Diabetes Prediction Dataset](https://www.kaggle.com/datasets/iammustafatz/diabetes-prediction-dataset)

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
import numpy as np
Q1 = np.percentile(data['Age'], 25)
Q3 = np.percentile(data['Age'], 75)

# Calculate IQR
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
filtered_data = data['Age'][(data['Age'] >= lower_bound) & (data['Age'] <= upper_bound)]
```

```python
filtered_data
```

```python
# Plotting the histogram with distribution curve
plt.figure(figsize=(10, 6))
sns.histplot(filtered_data, bins=20, kde=True, color='skyblue')

plt.title("Age Distribution with Density Curve", fontsize=12)
plt.xlabel("Age (Years)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)

plt.show()
```

```python
# Plot boxplot without outliers
plt.boxplot(filtered_data, vert=False)
plt.title('Box Plot of Age')
plt.ylabel('Age')
plt.show()
```

```python
# Drop rows where 'Age' is an outlier
data.drop(data[(data['Age'] < lower_bound) | (data['Age'] > upper_bound)].index, inplace=True)
data
```

#### Oversampling the Minority Class (Using SMOTE)

```python
X=data.drop('Diabetes',axis=1)  # Features
Y=data['Diabetes'] # Target
```

```python
X
```

```python
Y
```

```python
from imblearn.over_sampling import SMOTE

# Apply SMOTE
smote = SMOTE(sampling_strategy='auto', random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, Y)

# Create a new balanced DataFrame
data_balance = pd.DataFrame(X_resampled, columns=X.columns)
data_balance['Diabetes'] = y_resampled

data_balance['Diabetes'].value_counts()
```

```python
# Import libraries
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import cross_val_score
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create an array to store the average accuracy for different n_estimators
n_estimators_range = range(50, 301, 10)  # Test estimators from 50 to 200 with a step of 10
cv_scores = []  # To store cross-validation scores

# Step 2: Iterate over different n_estimators and use cross-validation
for n_estimators in n_estimators_range:
    catboost = CatBoostClassifier(n_estimators=n_estimators, random_seed=42, silent=True)  # Set silent=True to suppress training output
    scores = cross_val_score(catboost, X_train, y_train, cv=5, scoring='accuracy')
    cv_scores.append(scores.mean())  # Store the mean of cross-validation scores

# Step 3: Plot the results to find the optimum n_estimators
plt.plot(n_estimators_range, cv_scores, marker='o')
plt.xlabel('Number of Estimators')
plt.ylabel('Cross-Validated Accuracy')
plt.title('Optimum Number of Estimators for CatBoost')
plt.show()

# Step 4: Choose the n_estimators with the highest cross-validation score
optimal_n_estimators = n_estimators_range[np.argmax(cv_scores)]
print(f"Optimal number of estimators: {optimal_n_estimators}")

# Step 5: Train the CatBoost classifier with the optimal n_estimators
catboost_optimal = CatBoostClassifier(n_estimators=optimal_n_estimators, random_seed=42, silent=True)
catboost_optimal.fit(X_train, y_train)

# Step 6: Make predictions on the test set
y_pred = catboost_optimal.predict(X_test)

# Step 7: Evaluate accuracy of the model
catboost_accuracy = accuracy_score(y_test, y_pred) * 100
print(f"CatBoost Accuracy: {catboost_accuracy:.2f}%")
```
