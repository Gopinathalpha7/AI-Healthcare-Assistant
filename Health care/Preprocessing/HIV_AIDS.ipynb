# Problem Statement:
--------------------
HIV/AIDS remains a major public health challenge globally, impacting millions of lives and necessitating timely diagnosis and effective intervention to control its spread. Early identification of individuals at risk of HIV infection can be critical for prevention and treatment efforts, enhancing the quality of life and reducing transmission rates. This project aims to develop a machine learning model to predict the likelihood of HIV infection based on demographic, behavioral, and educational factors gathered from hospital data.

### Key Objectives:
1. Identify key demographic and behavioral factors associated with HIV infection risk.
2. Develop and train a predictive model using the provided dataset to classify individuals as HIV-positive or HIV-negative.
3. Evaluate model performance using metrics such as accuracy, precision, recall, F1-score, etc., to ensure reliability.
4. Provide insights into influential features impacting the model’s predictions to aid healthcare providers in targeted awareness and intervention efforts.

### Dataset link: [HIV Prediction Dataset](https://www.kaggle.com/datasets/ishigamisenku10/hiv-prediction)

### Approach:
1. **Data Preprocessing**: Clean and preprocess the dataset by handling missing values, normalizing/standardizing numerical features, and encoding categorical variables.
2. **Exploratory Data Analysis (EDA)**: Perform EDA to understand feature distributions and potential correlations with the target variable (HIV status).
3. **Feature Selection**: Identify the most relevant features through correlation analysis or feature importance techniques from machine learning models.
4. **Model Development**: Implement machine learning algorithms such as Logistic Regression, Decision Trees, Random Forest, or Gradient Boosting to predict HIV status.
5. **Model Evaluation**: Use cross-validation and performance metrics such as accuracy, precision, recall, F1-score, and ROC-AUC to assess the models.
6. **Model Interpretation**: Analyze feature importance to understand the model’s decision-making process and provide interpretable results that can assist healthcare providers.

### Results:
The anticipated outcome is a reliable machine learning model capable of accurately predicting HIV status, along with insights into key behavioral and demographic factors associated with HIV risk. This model aims to support early identification of individuals at risk, enabling targeted education and preventive measures to improve community health outcomes.

# About Dataset
--------------------

**Data Description**

The HIV Prediction dataset includes demographic, behavioral, and educational features that could aid in predicting HIV status. The data captures information from individuals regarding their marital status, sexual orientation, drug use, and history of STD and HIV testing.

### Features:

1. **Age** | Age of the individual | `Age` | int
2. **Marital Status** | Marital status | `Marital Staus` | categorical
   - Unmarried, Married, Other
3. **STD** | History of sexually transmitted diseases | `STD` | boolean
4. **Educational Background** | Education level | `Educational Background` | categorical
   - College Degree, Senior High School, Other
5. **HIV Test in Past Year** | Whether tested for HIV in the past year | `HIV TEST IN PAST YEAR` | boolean
6. **AIDS Education** | Participation in AIDS education | `AIDS education` | boolean
7. **Places of Seeking Sex Partners** | Location for seeking partners | `Places of seeking sex partners` | categorical
   - Internet, Bar, Other
8. **Sexual Orientation** | Orientation of the individual | `SEXUAL ORIENTATION` | categorical
   - Heterosexual, Homosexual, Other
9. **Drug-taking** | History of drug use | `Drug- taking` | boolean



**Target Variable**: HIV status of the individual
10. **Result** | HIV status of the individual | `result` | categorical
   - POSITIVE
   - NEGATIVE

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

### Model Selection.

#### 1. **Logistic Regression**

```python
# Import library
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Initialize
log_reg = LogisticRegression()

# Fit the model
log_reg.fit(X_train, y_train)

# Predictions
y_pred = log_reg.predict(X_test)

# Accuracy of model in %
logreg_accuracy = accuracy_score(y_test, y_pred) * 100
print(f"Logistic Regression Accuracy: {logreg_accuracy:.2f}%")
```

#### 2. **K-Nearest Neighbors (KNN)**

```python
# Import library
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Initialize
knn = KNeighborsClassifier()

# Fit the model
knn.fit(X_train, y_train)

# Predictions
y_pred = knn.predict(X_test)

# Accuracy of model in %
knn_accuracy = accuracy_score(y_test, y_pred) * 100
print(f"K-Nearest Neighbors Accuracy: {knn_accuracy:.2f}%")
```

#### 3. **Support Vector Machines (SVM)**

```python
# Import library
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Initialize
svm = SVC()

# Fit the model
svm.fit(X_train, y_train)

# Predictions
y_pred = svm.predict(X_test)

# Accuracy of model in %
svm_accuracy = accuracy_score(y_test, y_pred) * 100
print(f"Support Vector Machine Accuracy: {svm_accuracy:.2f}%")
```

#### 4. **Decision Tree Classifier**

```python
# Import library
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Initialize
decision_tree = DecisionTreeClassifier()

# Fit the model
decision_tree.fit(X_train, y_train)

# Predictions
y_pred = decision_tree.predict(X_test)

# Accuracy of model in %
dt_accuracy = accuracy_score(y_test, y_pred) * 100
print(f"Decision Tree Accuracy: {dt_accuracy:.2f}%")
```

#### 5. **Random Forest Classifier**

```python
# Import library
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Initialize
random_forest = RandomForestClassifier()

# Fit the model
random_forest.fit(X_train, y_train)

# Predictions
y_pred = random_forest.predict(X_test)

# Accuracy of model in %
rf_accuracy = accuracy_score(y_test, y_pred) * 100
print(f"Random Forest Accuracy: {rf_accuracy:.2f}%")
```

#### 6. **Naive Bayes Classifier**


```python
# Import library
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# Initialize
naive_bayes = GaussianNB()

# Fit the model
naive_bayes.fit(X_train, y_train)

# Predictions
y_pred = naive_bayes.predict(X_test)

# Accuracy of model in %
nb_accuracy = accuracy_score(y_test, y_pred) * 100
print(f"Naive Bayes Accuracy: {nb_accuracy:.2f}%")
```

#### 7. **Gradient Boosting Classifier**

```python
# Import library
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score

# Initialize
gradient_boost = GradientBoostingClassifier()

# Fit the model
gradient_boost.fit(X_train, y_train)

# Predictions
y_pred = gradient_boost.predict(X_test)

# Accuracy of model in %
gb_accuracy = accuracy_score(y_test, y_pred) * 100
print(f"Gradient Boosting Accuracy: {gb_accuracy:.2f}%")
```

### Model Building and Fine turning (LightGBM)

```python
import numpy as np
import matplotlib.pyplot as plt
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score

# Step 1: Define the range of boosting rounds to test
n_estimators_range = range(10, 210, 10)  # Testing n_estimators from 10 to 200
cv_scores = []  # To store cross-validation scores

# Step 2: Iterate over different numbers of boosting rounds (n_estimators)
for n_estimators in n_estimators_range:
    lgbm = LGBMClassifier(n_estimators=n_estimators, random_state=42)
    scores = cross_val_score(lgbm, X_train, y_train, cv=5, scoring='accuracy')
    cv_scores.append(np.mean(scores))  # Store the mean cross-validation score

# Step 3: Plot the results to find the optimum number of boosting rounds
plt.plot(n_estimators_range, cv_scores, marker='o')
plt.xlabel('Number of Boosting Rounds (n_estimators)')
plt.ylabel('Cross-Validated Accuracy')
plt.title('Optimum Number of Boosting Rounds for LightGBM')
plt.show()

# Step 4: Choose the number of boosting rounds with the highest cross-validation score
optimal_n_estimators = n_estimators_range[np.argmax(cv_scores)]
print(f"Optimal number of boosting rounds: {optimal_n_estimators}")

# Step 5: Train the LightGBM classifier with the optimal number of boosting rounds
lgbm_optimal = LGBMClassifier(n_estimators=optimal_n_estimators, random_state=42)
lgbm_optimal.fit(X_train, y_train)

# Step 6: Evaluate the model on the test set
y_pred = lgbm_optimal.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Test set accuracy: {accuracy*100:.2f}")
```
